```python
import numpy as np
import pandas as pd
from warnings import filterwarnings
filterwarnings('ignore')
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import seaborn as sns
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm
```

```python
# 데이터 
df_2020 = pd.read_csv("해상물류(트레드링스)_데이터파일/terminal_schedule_2020.csv", encoding='cp949')
df_2020.head()
df_2020.shape

df_2019 = pd.read_csv("해상물류(트레드링스)_데이터파일/terminal_schedule_2019.csv", encoding='cp949')
df_2019.head()
df_2019.shape

df_2018 = pd.read_csv("해상물류(트레드링스)_데이터파일/terminal_schedule_2018.csv", encoding='cp949')
df_2018.head()
df_2018.shape

df_2017 = pd.read_csv("해상물류(트레드링스)_데이터파일/terminal_schedule_2017.csv", encoding='cp949')
df_2017.head()
df_2017.shape
```

```python
df = pd.concat([df_2019, df_2020])
df = df.reset_index()
df.drop('index', axis=1, inplace=True)
df.head(5)
df.shape
```

```python
# 선박 작업 수량, 정보 수집일시 column 제거
df.drop(['Shit 작업 수량'],axis=1, inplace=True)
```

```python
df.head()
```

```python
# 문자열 전환
df['접안 일시'] = df['접안 일시'].astype(str)
df['출발 일시'] = df['출발 일시'].astype(str)
```

```python
# 소수점 자리 제거
df['접안 일시'] = df['접안 일시'].str.slice(0,14)
df['출발 일시'] = df['출발 일시'].str.slice(0,14)
```

```python
# 날짜 형식으로 문자열 수정
df['접안 일시'] = df['접안 일시'].str[0:4]+'-'+ df['접안 일시'].str[4:6]+'-'+df['접안 일시'].str[6:8] +' '+df['접안 일시'].str[8:10]+':'+df['접안 일시'].str[10:12]+':'+ df['접안 일시'].str[12:14]

df['출발 일시'] = df['출발 일시'].str[0:4]+'-'+ df['출발 일시'].str[4:6]+'-'+df['출발 일시'].str[6:8] +' '+df['출발 일시'].str[8:10]+':'+df['출발 일시'].str[10:12]+':'+ df['출발 일시'].str[12:14]
```

```python
# 날짜 포멧에 맞지 않는 데이터 삭제
df = df[df['접안 일시'].str.len() == 19]
df = df[df['출발 일시'].str.len() == 19]
```

```python
df.head()
```

```python
# 날짜형 데이터형 변환
df['접안 일시'] = pd.to_datetime(df['접안 일시'])
df['출발 일시'] = pd.to_datetime(df['출발 일시'])
```

```python
df.head()
type(df['접안 일시'][0])
```

```python
df.isnull().sum()
```

```python
# 작업수량 0인 결측치 제거
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.head()
df.shape
```

```python
# 잔류시간 행 추가
df['잔류 시간'] = df['출발 일시'] - df['접안 일시']
df['잔류 시간'].describe()
```

```python
# 도착이 출발보다 더 빠른 오류 데이터 탐색
df[(df['접안 일시'] >= df['출발 일시'])== True].count()
```

```python
df.shape
```

```python
# 삭제
df.drop(df.loc[df['출발 일시'] <= df['접안 일시']].index, inplace=True)
df.shape
```

```python
# 작업 수량 데이터 분포 탐색
df['선적 작업 수량'].describe()
df['양하 작업 수량'].describe()
```

```python
df.shape
```

```python
# 작업 수량이 0 이하인 column 제거
df.drop(df.loc[df['선적 작업 수량']<=0].index, inplace=True)
df.drop(df.loc[df['양하 작업 수량']<=0].index, inplace=True)

df.shape
```

```python
# 정보수집 일시 날짜형 변환
df['정보 수집일시'] = pd.to_datetime(df['정보 수집일시'])
df['정보 수집일시'] = pd.to_datetime(df['정보 수집일시'])
```

```python
# 정보 수집보다 늦게 출발한 경우 제외
df.shape
df.drop(df[df['출발 일시'] >= df['정보 수집일시']].index, inplace=True)
df.shape
```

```python
df.drop_duplicates(['터미널 코드','Call Year','선박 코드','Call No'], keep="last").head(10)
df.shape
```

```python
df['잔류 시간'] = df['잔류 시간'].dt.seconds + df['잔류 시간'].dt.days*86400
df.head()
```

```python
df.to_csv('terminal_schedule(2019~2020)_전처리완료(실제).csv', encoding='utf-8')
```
